```python
import numpy as np
import random
```

```python
def crear_tablero(filas=10, columnas=10):
    # Crea un tablero, relleno con espacios.
    return np.full((filas, columnas), " ")

def posicionar_barco(tablero, posiciones):
    # Coloca un barco (O) en las posiciones indicadas.
    for (fila, col) in posiciones:
        tablero[fila, col] = "O"

tablero = crear_tablero()
print("Tablero inicial:\n", tablero)
barco_2 = [(0,1), (1,1)]
barco_4 = [(1,3), (1,4), (1,5), (1,6)]
posicionar_barco(tablero, barco_2)
posicionar_barco(tablero, barco_4)
print("\nTablero con barcos colocados:\n", tablero)
```

```python
# Funcion disparar jugador
def disparar(tablero_rival,tablero_rival_2, fila, col):
   # Marca un disparo: X si había barco, - si era agua.
    if tablero_rival[fila, col] == "O":
        tablero_rival[fila, col] = "X"      # Aquí marca con una "X" el tablero que el jugador no ve.
        tablero_rival_2[fila, col] = "X"    # Aquí marca con una "X" el tablero que el jugador ve.
        print(f" ¡Impacto en ({fila}, {col})!")
    elif tablero_rival[fila, col] == " ":
        tablero_rival[fila, col] = "-"      # Aquí marca con una "-" el tablero que el jugador no ve.
        tablero_rival_2[fila, col] = "-"    # Aquí marca con una "-" el tablero que el jugador ve.
        print(f" Agua en ({fila}, {col}).")
    else:
        print(f" Ya se disparó en ({fila}, {col}).")
```

```python
disparar(tablero, tablero, 0, 1)
```

```python
# Recibir disparo
def disparo_rival(tablero):
    filas, cols = tablero.shape
    while True:
        fila = random.randint(0, filas - 1)
        col = random.randint(0, cols - 1)
        
        if tablero[fila, col] in ["O", " "]:  # solo dispara si no lo ha hecho antes
            if tablero[fila, col] == "O":
                tablero[fila, col] = "X"
                print(f"¡Impacto en ({fila}, {col})!")
                print(tablero)
            else:
                tablero[fila, col] = "-"
                print(f"Agua en ({fila}, {col}).")
                print(tablero)
            break  # sal del bucle tras disparar
```

```python
disparo_rival(tablero)
```
